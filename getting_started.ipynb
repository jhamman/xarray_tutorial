In this notebook, we'll just make sure we can import xarray and access the tutorial data. After that, you should head over to the `tutorial` directory to go through the rest of the material.

```python
# import xarray
import xarray as xr
```

```python
# load the air temperature dataset
# you can replace 'air_temperature' with `rasm` to get a different dataset

ds = xr.tutorial.load_dataset('air_temperature')
```

```python
ds
```

```python
# calculate the season means
avg_temp = ds['air'].groupby('time.season').mean('time')
```

```python
# make a quick plot
avg_temp.plot(col='season', robust=True)
```
